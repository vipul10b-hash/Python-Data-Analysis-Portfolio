```python
# ladder
a='10.9'
print('value of a',a)
print(type(a))
print('_'*100)
a=float(a)
print('value of a',a)
print(type(a))
print('_'*100)
a=int(a)
print('value of a',a)
print(type(a))
print('_'*100)
a=bool(a)
print('value of a',a)
print(type(a))
print('_'*100)
```
