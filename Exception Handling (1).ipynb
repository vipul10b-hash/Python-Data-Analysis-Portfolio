```python
try:
    n1=int(input('enter a number'))
    n2=int(input('enter a number'))
    print(f'division of {n1} and {n2} is {n1/n2}')
except ZeroDivisionError:
    print('denominator could not be zero')
except ValueError:
    print('please add valid integers only')
```

```python
# exception as e
try:
    n1=int(input('enter a number'))
    n2=int(input('enter a number'))
    print(f'division of {n1} and {n2} is {n1/n2}')
except Exception as e:
    print(e)
```

```python
# finally
try:
    n1=int(input('enter a number'))
    n2=int(input('enter a number'))
    print(f'division of {n1} and {n2} is {n1/n2}')
except Exception as e:
    print(e)
finally:
    print('end of code')
```
