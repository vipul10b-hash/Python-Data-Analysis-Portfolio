  # What is for loop?
  A for loop is used to repeat a block of code for each item in a sequence (like a list,string,tuple,etc.).
  # Basic Syntax:
  for variable in sequence:
  #code block to execute
  # Common Uses
  - iterating over elements in a list,tuple,set or dictionary
  - repeating code a fixed number of times using range()
  - processing string or files by line


```python
# usecase
print(1)
print(2)
print(3)
print(4)
```

# ITERABLES / SEQUENCE
### Definition:
An iterable in python is any object capable of returning its elements one at a time , allowing it to be used in a loop or with functions like iter() and next().
- list
- sets
- dictionaries
- str
- tuples

```python
# syntax
for variable in iterable:
    statements
```

```python
l1=['aqram','harsh','anushka','prince']
```

```python
# for loop
for i in l1:
    print(i)
```

```python
# sets
s1={'timeless','tum hi ho','excuses',295,'lover'}
s1
for i in s1:
    print(i)
    
```

```python
# tuples
t1=('gulab jamun','kaju katli','gulab jamun','gevar','laddu','imarti')
```

```python
for i in t1:
    print(i)
```

```python
# dictionaries
dict1={'name':'varsha',
       'course':'data analytics',
       'voice pitch':'very low',
       'age':21}
    
```

```python
for i,j in dict1.items():
    print(i,'\t\t',j)
```

```python
dict1.items()
```

```python
# what if 1-100 or 1-1000,1-100000
# range function
# range(start,end,stepsize)
```

```python
# typecasting
list(range(1,11))
```

```python
for i in range(0,101,10):
    
    #last value is step size in this case this is 10
    print(i)
```

```python
for i in range(0,101,1):
    print(i)
    
```

```python
# 1 parameter
for i in range(5):
    print(i)
```

```python
# iter function
l1=[20,31,42,34,55]
m=iter(l1)
```

```python
# next function
print(next(m))
print(next(m))
print(next(m))
print(next(m))
print(next(m))
```

```python
# pass
# continue
# break
for i in range (1,101):
    # future reference
    pass 
```

```python
# break
for i in range(1,6):
    print(i)
    if i==3:
        break
```

```python
# continue
for i in range(1,6):
    if i==3:
        continue
    print(i)
    print('-'*20)
```
