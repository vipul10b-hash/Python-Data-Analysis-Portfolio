

### 1. **Required Arguments**

Arguments that must be passed to a function in the correct positional order.

### 2. **Keyword Arguments**

Arguments passed by explicitly specifying the parameter name, allowing order flexibility.


### 3. **Default Arguments**

Arguments that assume a default value if no value is provided during the function call.


### 4. **Arbitrary Arguments (`*args`)**

Used to pass a variable number of **positional** arguments to a function.


### 5. **Arbitrary Keyword Arguments (`**kwargs`)**

Used to pass a variable number of **keyword** arguments to a function.


# Required Arguments
Required arguments are parameters that must be passed to the function in the correct positional order.If not provided,Python will raise a TypeError.

```python
#required arguements
def sum(x,y):
    return x+y
```

# Default Arguments
Default arguements are parameters that assume a default value if a value is not provided in the function call.They are specified by assigning a value to the parameter in the function defination.

```python
def sum(x=0,y=0):
    return x+y
```

```python
sum()
sum(234,34)
```

# Keyword Arguments
keyword arguement are passed to a functions by explicity naming the parameters and assigning them values in the function call. This allows you to specify values in the function call.This allows you to specify values for only some parameters while relying on default values of others , and to pass arguments in a different order.

```python
def call(x=10,y=20):
    # return x,y
    print('x=',x)
    print('y=',y)
    
```

```python
call(y=30)
```

```python
#reordering 
call(y=0,x=100)
```

```python
greetings(fname='vipul',occup='data science')
```

# Variable-length arguments
Variable length arguments allow a function to accept an arbitary numbers of arguments.They have defined using *args for naon-keyword arguments and **kwargs for keyword arguments

# *args
*args allows a function to accept any numbers of positional arguements.The arguements are passed as tuple.(iterable)


```python
#args
# *
def funct1(*args):
    print(args)
    print(type(args))
```

```python
funct1(13,25,5416,656,6161,61,51,15416,5162)
```

```python
#mean
def mean(*args):
    sum=0
    for i in args:
        sum+=i
    #print(sum)
    return sum/len(args)
```

```python
mean(12,6265,622,65692,5,25,152,626,25562,895)
```

```python
mean(10,20,30,40,50,60,70)
```

# **kwargs
**kwargs allows a function to accept numbers of keyword arguements.The arguements are passed a dictionary

```python
def funct2(**kwargs):
    print(kwargs)
    print(type(kwargs))
```

```python
funct2(name='vipul',desig='Data scientist') 
```

```python
def info(**kwargs):
    # print()
    global var1
    var1 = kwargs
    for i,j in kwargs.items():
        print(i,'\t',j)
```

```python
info(name='vipul',age=19,crs='DS')
```
