# Lambda
A lambda function is a small anonymous functions defined with the lambda keyword.It can have any number of arguments but only one expression.It's often used for a short,simple function that are needed temporarily.

```python
# aam zindagi
def sum(x,y):
    return(x+y)
```

```python
# mentos zindagi
# lambda function
sum= lambda x,y : x+y
```

```python
sum(109,20)
```

```python
# single arguements
# cube
cube = lambda x:x**3
```

```python
cube(3)
```

```python
# 3-4
mean_of_3=lambda x,y,z: (x+y+z)/3
```

```python
mean_of_3(10,12,236)
```

# Map Functions
The map function applies a given function to all items in an input list (or any other iterable) and return a map objec(which is an iterator).You can convert object into list or other data structure is needed.


```python
# aam zindagi
l1=[1,2,3,4,5,56,783,23,23]
square= lambda x:x**2
# square(l1)
n1=[]
for i in l1:
    # print(square(i))
    n1.append(square(i))

n1
```

```python
list(map(square,l1)) #mentos zindagi
```

```python
l1=[2,5,7,10] #--> cube
list(map(lambda x:x**3,l1))
```

```python
# even odd
list(map(lambda x: 'even' if x%2==0 else 'odd',l1))
```

#### Reduce 

The reduce function is part of the functools module. It applies a rolling computation a sequential pairs of values in an iterable. It performs a repetative operation over the pairs of the iterable.

#### syntax

reduce(function, iterable)

```python
# reduce
from functools import reduce
```

```python
# aam zindagi
l1=[1,2,3,4,5,6]
sum=0
for i in l1:
    sum+=i
sum
```

```python
# mentos zindagi
reduce(lambda x,y:x+y,l1)
```

# Filter

The filter constructs an iterator from elements of an iterable if an iterable for which a funtion returns true. it Filters out the elements that do not satisfy the condition provided by the function.

- filter(function, iterable)

```python
# filter
l1 = list(range(0,50,3))
l1
```

```python
list(filter(lambda x: True if x>25 else False,l1))
```

# Recursion
Recursion in Python is a programming technique where a function calls itself directly or indirectly to solve a problem. This approach is particularly useful for problems that can be naturally broken down into smaller, self-similar subproblem

```python
# factorial
# 5! = 5*4*3*2*1
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
factorial(5)
```

```python
# 1! = 1
# 0! = 1
```

```python
n = int(input('enter a number'))
prod = 1
for i in range(1,n+1):
    prod*=i
prod
```
