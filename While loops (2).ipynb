# while Loops in Python 
The while loop in Python is used to execute a block of code repeatedly as long as a given condition is True.
# while loops 

# syntax

while condition:
    statements 

```python
# basic code
#1-5
choice=1
while choice<6: #true
    print(choice)
    choice+=1
```

```python
#reverse loop 
choice=5
while choice>0: #true
    print(choice)
    choice-=1
```

# list comprehensions
List comphrehensions provide a concise and efficient way to create new lists in python. they offer a more readable and often faster alternative to traditional for loops for list creation and manipulations. 


new_list=[expression for item in iterable if condition]

```python
l1=[1,2,3,4,5,6,7]
n1=[]
for i in l1:
    n1.append(i**2)
n1
```

```python
[x**2 for x in l1]
```

```python
# if condition
# filter
print(l1)
[x for x in l1 if x%2==0]
```

```python
[operation for variable in iterable if condition]
```
